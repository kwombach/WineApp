```python
import matplotlib.pyplot as plt
from price_clouds import word_clouds
from wordcloud import WordCloud, ImageColorGenerator
import pandas as pd
import numpy as np
from os import path, getcwd
from PIL import Image
import numpy as np
```

```python
info = word_clouds('wine', 'review_wineId_int')
```

```python
def wc_two_wines(wineid1, wineid2):
    
    x = [data[0] for data in info]
    wine1 = info[x.index(wineid1)]
    wine2 = info[x.index(wineid2)]
    
    words1 = wine1[1]
    wordcloud1 = WordCloud(max_font_size = 75).generate_from_frequencies(frequencies=words1.items())


    plt.imshow(wordcloud1, interpolation='bilinear')
    plt.axis("off")
    plt.savefig('wine1.png')
    plt.show()
    
    
    words2 = wine2[1]
    wordcloud2 = WordCloud(max_font_size = 75).generate_from_frequencies(frequencies=words2.items())

    plt.imshow(wordcloud2, interpolation='bilinear')
    plt.axis("off")
    plt.savefig('wine2.png')
    plt.show()


```

```python
wc_two_wines(23,15269)
```
