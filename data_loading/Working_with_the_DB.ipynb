### Make a query using above and then you can dump the query to a pandas df if you want with this style syntax. Probably don't want to limit it though.

```python
import sqlite3
import csv
import pandas as pd

con = sqlite3.connect('wineapp.db')
cur = con.cursor()
sql = '''SELECT *
           FROM wines
          LIMIT 100'''
wine_df = pd.read_sql(sql, con)
con.commit()

wine_df
```
