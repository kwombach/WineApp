```python
import time
import requests
from bs4 import BeautifulSoup
import random as random
import numpy as np
from selenium import webdriver

browser = webdriver.Chrome('/Users/bgodd/Desktop/Brown_Data_Science/data1030/FinalProject/chromedriver')  

wine_list = [i for i in range(100000)]
random.shuffle(wine_list)

for wine_no in wine_list:
    wine_no_str = str(wine_no)
    time.sleep(3+np.random.uniform(0,1))
    
    browser.get('https://www.cellartracker.com/wine.asp?iWine='+wine_no_str)
    if browser.title != 'Wine Not Found, - CellarTracker':
        time.sleep(3+np.random.uniform(0,1))
        main = browser.page_source
        browser.get('https://www.cellartracker.com/notes.asp?iWine='+wine_no_str)
        time.sleep(3+np.random.uniform(0,1))
        if browser.title != 'Wine Not Found, - CellarTracker':
            tasting = browser.page_source
            with open('scraped/main_'+wine_no_str+'.html', 'w') as file:
                file.write(str(main))

            with open('scraped/tasting_'+wine_no_str+'.html', 'w') as file:
                file.write(str(tasting))

            scraped_wines  = open('readWines.txt', 'a')
            scraped_wines.write(wine_no_str+'\n') 
            scraped_wines.close()
        else:
            scraped_wines  = open('readWines.txt', 'a')
            scraped_wines.write(wine_no_str+'_X_Unread\n') 
            scraped_wines.close()
    else:
        scraped_wines  = open('readWines.txt', 'a')
        scraped_wines.write(wine_no_str+'_X_Unread\n') 
        scraped_wines.close()

browser.quit()

# print(main)
# # print(r.text)

# with open("output1.html", "w") as file:
#     file.write(str(soup))
    

# soup = BeautifulSoup(r.text, 'lxml')
```

```python
import json
from pprint import pprint
import pandas as pd

wine_list = []

# DataFrame.from_dict(data, orient='columns', dtype=None)[source]

for file in range(122):
    file+=1
    data = json.load(open('data_lcbo_'+str(file)+'.txt'))

    for item in data['result']:
        if item['primary_category'] == 'Wine':
                #print (str(item['name'])+', '+str(item['tasting_note']))
                wine_list.append((str(item['name']),str(item['tasting_note'])))


pprint(data['result'][1])
```

```python
# wine_list
labels = ['Name', 'Tasting_Notes']
df = pd.DataFrame.from_records(wine_list, columns=labels)
df=df.loc[df['Tasting_Notes'] != 'None']
# pd.DataFrame.from_dict(wine_list,dtype=None)
df.to_csv('LCBO.csv')
```
