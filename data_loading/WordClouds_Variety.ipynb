```python
import matplotlib.pyplot as plt
from price_clouds import word_clouds
from wordcloud import WordCloud, ImageColorGenerator
import pandas as pd
import numpy as np
from os import path, getcwd
from PIL import Image
import numpy as np
```

```python
data = word_clouds('variant', 'wine_variant')
```

```python
# Cabernet Sauvignon : data[19]
# Chardonnay : data[32]
# Malbec : data[93]
# Merlot : data[111]
# Pinot Noir : data[171]
# Riesling : data[187]
# Savingnon Blanc : data[202]
# Zinfindel : data[261]
```

```python
# Cabernet Savignon Word Cloud

words = data[19][1]

d = getcwd()
mask = np.array(Image.open(path.join(d, "Glass.png")))
wordcloud = WordCloud(background_color="white", max_words=1000, mask=mask,
               max_font_size=90, random_state=42)
wordcloud.generate_from_frequencies(frequencies=words.items())

image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[15,15])
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.savefig('Cabernet.png')
_=plt.show()
```
